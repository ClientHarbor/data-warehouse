# Running Function through Threading - Streamlining workers with CPU

```python
def process_hcad(hcad):
    aprsls = harris_county_appraisals(hcad_id=hcad["id"])
    if aprsls:
        for aprsl in aprsls:
            year = list(aprsl.keys())[0]
            if year in {"2020", "2021", "2022", "2023", "2024"}:
                df.loc[hcad["index"], f"{year}_Appraisal"] = aprsl[year]
    print("Processed: ", hcad["index"])
```

```python
from concurrent.futures import ThreadPoolExecutor

num_workers = 10
with ThreadPoolExecutor(max_workers=num_workers) as executor:
    executor.map(process_hcad, result[10760:])
```
